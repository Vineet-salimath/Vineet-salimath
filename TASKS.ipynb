Problem 1

```python
class SmartLight:
    def __init__(self, light_name, action):
        self.light_name = light_name
        self.action = action
    def display_status(self):
        print(f"{self.light_name} is {self.action}")

bedroom_light = SmartLight("Bedroom Light", "ON")
livingroom_light = SmartLight("LivingRoom Light", "OFF")
bedroom_light.display_status()
livingroom_light.display_status()
print()
```

Problem 2

```python
class Employee:
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department
    def display_id_card(self):
        print("Employee ID Card")
        print(f"Name: {self.name}")
        print(f"ID: {self.emp_id}")
        print(f"Department: {self.department}")

employee_one = Employee("Rahul", "EMP102", "IT")
employee_one.display_id_card()
print()
```

Problem 3

```python
class ContactRecord:
    def __init__(self, contact_name, phone_number):
        self.contact_name = contact_name
        self.phone_number = phone_number
    def display_contact(self):
        print("Contact Saved")
        print(f"Name: {self.contact_name}")
        print(f"Phone: {self.phone_number}")

my_contact = ContactRecord("Vineet", "9876543210")
my_contact.display_contact()
print()
```

```python
class Product:
    def __init__(self, product_name, price):
        self.product_name = product_name
        self.price = price
    def generate_price_tag(self):
        print(f"Product: {self.product_name}")
        print(f"Price: {self.price}")

item = Product("Headphones", 2499)
item.generate_price_tag()
print()
```

Problem 5

```python
class Movie:
    def __init__(self, movie_name, rating):
        self.movie_name = movie_name
        self.rating = rating

    def display_rating(self):
        print(f"Movie: {self.movie_name}")
        print(f"Rating: {self.rating}")

movie_review = Movie("Inception", 4.8)
movie_review.display_rating()
print()
```

```python
class DeliveryManager:
    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address
    def print_delivery_details(self):
        print("Delivery Details")
        print(f"Customer: {self.customer_name}")
        print(f"Address: {self.address}")

delivery = DeliveryManager("Suman", "Hyderabad")
delivery.print_delivery_details()
print()
```
