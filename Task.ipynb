```python
def analyze_results(marks):
    pass_count = 0
    fail_count = 0

    for i in marks:
        if i >= 50:
            pass_count += 1
        else:
            fail_count += 1

    print(f"Total Students Passed: {pass_count}")
    print(f"Total Students Failed: {fail_count}")

student_marks = [45, 78, 90, 33, 60]  #example
analyze_results(student_marks)
```

```python
def clean_names(names):
    cleaned = []
    for i in names:
        cleaned_name = i.strip().lower()  # remove spaces and convert to lowercase
        cleaned.append(cleaned_name)
    return cleaned
names_list = [" Peter ", "VIJAY", " Vineet "] #example
cleaned_list = clean_names(names_list)
print("Cleaned Names:", cleaned_list)

```

```python
def message_analysis(messages):
    for msg in messages:
        length = len(msg)
        if length > 10:
            flag = " big message"
        else:
            flag = ""
        print("Message: '" + msg + "' | Length: " + str(length) + flag)

messages_list = ["Hi", "Welcome to the platform", "OK"] #example
message_analysis(messages_list)
```

```python
def count_messages(logs, msg_type):
    count = logs.count(msg_type)
    print(msg_type + " messages:", count)

system_logs = ["INFO", "ERROR", "WARNING", "ERROR"]   #example
count_messages(system_logs, "ERROR")
```

```python
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}
all_scores = list(employees.values())
highest_score = max(all_scores)
top_performers = []
for name, score in employees.items():
    if score == highest_score:
        top_performers.append(name)

names_string = ", ".join(top_performers)
print(f"Top Performers Eligible for Bonus: {names_string} (Score: {highest_score})")
print("\n")
```

Problem 2

```python
search_query = "Buy mobile phone buy phone online"
query_lower = search_query.lower()
#  Split the sentence into individual words
words = query_lower.split()
# Count frequency of each word using a dictionary
word_counts = {}
for word in words:
    if word in word_counts:
        word_counts[word] += 1
    else:
        word_counts[word] = 1
duplicates = {}
for word, count in word_counts.items():
    if count > 1:
        duplicates[word] = count
print(duplicates)
print("\n")
```

Problem 4

```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"]

domain_counts = {}
total_emails = len(emails)

# Extract domain from each email
for email in emails:
    # Split the email at '@' and take the second part
    parts = email.split('@')
    domain = parts[1]
    if domain in domain_counts:
        domain_counts[domain] += 1
    else:
        domain_counts[domain] = 1
for domain, count in domain_counts.items():
    percentage = (count / total_emails) * 100
    print(f"{domain}: {int(percentage)}%")

print("\n")
```

Problem 5

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]
total_sales = sum(sales)
number_of_days = len(sales)
average_sales = total_sales / number_of_days

threshold = average_sales * 1.30

# Loop through sales to find spikes
for day, sale_value in enumerate(sales, start=1):
    if sale_value > threshold:
        print(f"Day {day}: {sale_value}")

print("\n")
```
