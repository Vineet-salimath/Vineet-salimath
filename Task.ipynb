```python
def message_analysis(messages):
    for msg in messages:
        length = len(msg)
        if length > 10:
            flag = " big message"
        else:
            flag = ""
        print("Message: '" + msg + "' | Length: " + str(length) + flag)

messages_list = ["Hi", "Welcome to the platform", "OK"] #example
message_analysis(messages_list)
```

```python
def count_messages(logs, msg_type):
    count = logs.count(msg_type)
    print(msg_type + " messages:", count)

system_logs = ["INFO", "ERROR", "WARNING", "ERROR"]   #example
count_messages(system_logs, "ERROR")
```

```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"]

domain_counts = {}
total_emails = len(emails)

# Extract domain from each email
for email in emails:
    # Split the email at '@' and take the second part
    parts = email.split('@')
    domain = parts[1]
    if domain in domain_counts:
        domain_counts[domain] += 1
    else:
        domain_counts[domain] = 1
for domain, count in domain_counts.items():
    percentage = (count / total_emails) * 100
    print(f"{domain}: {int(percentage)}%")

print("\n")
```

```python
def validate_recharge():
    valid_plans = [50,199, 299, 399, 599]
    while True:
        try:
            amount = int(input("Enter your recharge amount (₹): "))
        except ValueError:
            print("Please enter a valid number.")
            continue


        if amount < 50:
            print("Failed: Minimum recharge amount is ₹50. Please try again.")

        elif amount not in valid_plans:
            print(f"Failed: ₹{amount} is not a valid plan. Choose from {valid_plans}.")
        else:
            print("Success! Your recharge is being processed.")
            break

validate_recharge()
```

```python
def process_result(marks_list):
    total_score = 0
    for mark in marks_list:
        total_score = total_score + mark
    number_of_subjects = len(marks_list)
    average_score = total_score / number_of_subjects

    print(f"Student Average: {average_score}")
    if average_score >= 50:
        return "Pass"
    else:
        return "Fail"
student_marks = [45, 60, 55, 40, 70]

print("Final Result:", process_result(student_marks))
```

```python
def check_eligibility(attendance_record):
    present_count = 0
    for status in attendance_record:
        if status == "P":
            present_count += 1


    total_days = len(attendance_record)
    attendance_percentage = (present_count / total_days) * 100

    print(f"Total Attendance: {attendance_percentage}%")


    if attendance_percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"


employee_attendance = ["P", "P", "A", "P", "P", "A", "P", "P", "P", "P"]

print("Employee Status:", check_eligibility(employee_attendance))
```

```python
def check_password_strength(password):
    if len(password) < 8:
        return "Weak: Password must be at least 8 characters long."
    has_digit = False
    has_special = False


    for char in password:
        if char.isdigit():
            has_digit = True
        if char in "@#$":
            has_special = True


    if has_digit and has_special:
        return "Strong: Password is secure!"
    elif not has_digit:
        return "Weak: Password needs at least one number."
    else:
        return "Weak: Password needs at least one special character (@, #, or $)."

#print(check_password_strength("password123"))  # Weak
print(check_password_strength("SecureP@ssw0rd"))  # Strong
```
